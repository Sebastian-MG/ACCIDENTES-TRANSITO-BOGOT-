```python
# importar librerias necesarias
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
#leer datos
datos_accidentes =  "2015_2019_siniestralidad_vial.xlsx"
# leer datos de accidentes hoja siniestros
accidentes_siniestros= pd.read_excel("2015_2019_siniestralidad_vial.xlsx", sheet_name="SINIESTROS")
```

```python
accidentes_siniestros.info()
```

```python
# se realiza histogramas
# variable categórica Gravedad
wplot = sns.countplot(data=accidentes_siniestros,x="GRAVEDAD")
wplot.set_xticklabels(wplot.get_xticklabels(), ha="right")
plt.savefig("Imágenes/Gravedad.jpg.jpg",bbox_inches='tight')
plt.show()
```

```python
# se realiza histogramas
# variable categórica clase
wplot = sns.countplot(data=accidentes_siniestros,x="CLASE")
wplot.set_xticklabels(wplot.get_xticklabels(), rotation=40, ha="right")
plt.savefig("Imágenes/Clase.jpg",bbox_inches='tight')
plt.show()
```
